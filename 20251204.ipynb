```python
# Iterable Object
l = [1,2,3]
print(type(l))
# print(next(l))
# TypeError: 'list' object is not an iterator

print("----------------")

# Iterator
my_iterator = iter(l) # Iterator로 변환
print(type(my_iterator))
# next() 메소드로 하나씩 꺼내기
print(next(my_iterator))
print(next(my_iterator))
print(next(my_iterator))
# print(next(my_iterator))
# StopIteration:
# 다음이 없어서 에러
```

```python
def bad_range(stop):
  nums = []
  i = 0
  while i < stop:
    nums.append(i)
    i += 1
  return nums

def good_range(stop):
  i = 0
  while i < stop:
    yield i
    i += 1

# 제너레이터 객체 생성
runner = good_range(3)
print(runner)
# 하나씩 꺼내기
print(next(runner))
print(next(runner))
print(next(runner))
```

```python
import sys

# 데이터를 미리 다 만듬
my_list = [i for i in range(10000)]
print(f"list 메모리 크기: {sys.getsizeof(my_list)} bytes")
# 생성 규칙만 정함
my_gen = (i for i in range(10000))
print(f"generator 메모리 크기: {sys.getsizeof(my_gen)} bytes")
```

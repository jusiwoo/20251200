```python
x = 'abcd'
y = x
print(f"{x}, {id(x)}")
print(f"{y}, {id(y)}")
y += 'e'
print(f"{x}, {id(x)}")
print(f"{y}, {id(y)}")
```

```python
x = (1, 2, 3)
y = x
print(f"{x}, {id(x)}")
print(f"{y}, {id(y)}")
y += (4,)
print(f"{x}, {id(x)}")
print(f"{y}, {id(y)}")
```

```python
x = [1,2,3]
y = x
print(f"{x}, {id(x)}")
print(f"{y}, {id(y)}")
y += [4,]
print(f"{x}, {id(x)}")
print(f"{y}, {id(y)}")
```

```python
print("--- Reference(단순 할당) ---")
x = [1,2,3]
y = x
print(f"{x}, {id(x)}")
print(f"{y}, {id(y)}")
print(y is x)
y[1] = 10
print(f"{x}, {id(x)}")
print(f"{y}, {id(y)}")
print(y is x)

print("--- shallow copy(얕은 복사) ---")
x = [[1,2,3],[4,5,6]]
y = x.copy() # 또는 x[:]
print(f"{x}, {id(x)}")
print(f"{y}, {id(y)}")
print(y is x)
print(y == x)
y[0][0] = 0
print(f"{x}, {id(x)}")
print(f"{y}, {id(y)}")
print(y is x)
print(y == x)

print("--- Deep Copy(깊은 복사) ---")
import copy
x = [[1,2,3],[4,5,6]]
y = copy.deepcopy(x)
print(f"{x}, {id(x)}")
print(f"{y}, {id(y)}")
print(y is x)
print(y == x)
y[0][0] = 0
print(f"{x}, {id(x)}")
print(f"{y}, {id(y)}")
print(y is x)
print(y == x)
```

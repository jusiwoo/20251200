```python
import random

class Character:
  def __init__(self, name, job, hp, attack):
    self.name = name
    self.job = job
    self.hp = hp
    self.attack = attack

  def __str__(self):
    return f"<이름: {self.name}> <직업: {self.job}> <체력: {self.hp}> <공격력: {self.attack}>"

  def __call__(self, target):
    damage = self.attack + random.randint(-5, 5)
    return damage

me = Character("홍길동", "전사", 110, 20)
print(me)
print(me("벌레"))
```
